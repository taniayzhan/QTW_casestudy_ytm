```r
txt = readLines("offline.final.trace.txt")
```

```r
#######
# This is taken from the R text book.  
# Basically, read in the line, split it by tokens and then assemle it into a data matrix.
#######
processLine = function(x)
{
tokens = strsplit(x, "[;=,]")[[1]]
if (length(tokens) == 10)
    return(NULL)
tmp = matrix(tokens[ - (1:10) ], , 4, byrow = TRUE)  #build a 4 column matrix mac,signal,scan,type
cbind(matrix(tokens[c(2, 4, 6:8, 10)], nrow(tmp), 6, #builds a 6 column matrix t,scanMac,x,y,z,angle
byrow = TRUE), tmp) #binds them together
}
```

```r
#first 3 lines are comments so lets look at line 4
txt[4]
```

```r
#split the line into tokens
strsplit(txt[4], "[;=,]")[[1]]
```

```r
#creating the 4 column matrix
tokens = strsplit(txt[4], "[;=,]")[[1]]
matrix(tokens[ - (1:10) ], , 4, byrow = TRUE)
```

```r
#creating the 6 column matrix
tmp = matrix(tokens[ - (1:10) ], , 4, byrow = TRUE)
matrix(tokens[c(2, 4, 6:8, 10)], nrow(tmp), 6,byrow = TRUE)
```

```r
#process the whole datafile
lines = txt[ substr(txt, 1, 1) != "#" ]
tmp = lapply(lines, processLine)
offline = as.data.frame(do.call("rbind", tmp),stringsAsFactors = FALSE)
```

```r
#process the online data file
txt = readLines("online.final.trace.txt")
lines = txt[ substr(txt, 1, 1) != "#" ]
tmp = lapply(lines, processLine)
online = as.data.frame(do.call("rbind", tmp),stringsAsFactors = FALSE)
```
